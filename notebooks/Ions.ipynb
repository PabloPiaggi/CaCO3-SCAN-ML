```python
filename = "ca-med/dump.water"
traj = aseio.iread(filename,format='lammps-dump-text',index=":2000:10")
myframes = []
#counter=0
#n_discard=1
#n_take=100
for atoms in tqdm(traj):
    #counter += 1
    #if (counter>n_discard):
    myframes.append(atoms)
    #if (counter>n_discard+n_take):
    #    break
```

```python
coord = []
for frame in tqdm(myframes):
    #print(frame.get_chemical_formula())
    i = neighbor_list('i', frame,{('H', 'Li'): 3.15})
    coord.append(np.bincount(i)[0])
print("Mean coordination is: ", np.mean(coord))
```

```python
coord = []
for frame in tqdm(myframes):
    #print(frame.get_chemical_formula())
    i = neighbor_list('i', frame,{('Li','Li'): 3.3, ('H','He'): 3, ('H','Be'): 5, ('He','Be'): 3, ('Be','Be'): 3})
    condition = (np.array(frame.get_chemical_symbols()) == "Li")
    try:
        coord.append(np.bincount(i)[condition])
    except:
        print("Problem")
    #print(frame.get_chemical_symbols(),condition)
    #print(np.bincount(i)[condition])
print("Mean coordination is: ", np.mean(coord)) # subtract the three carbonate oxygens
```

# g(r)

```python
rdf_CaO, rdf_CaH = [], []
for atoms in tqdm(myframes):
    ana = Analysis(atoms)
    #print(atoms.get_chemical_symbols())
    rmax = 10
    nbins = 80
    dummy = ana.get_rdf(rmax,nbins,return_dists=True,elements=(1,3))
    rdf_CaO.append(dummy[0][0])
    r = dummy[0][1]
    dummy = ana.get_rdf(rmax,nbins,return_dists=True,elements=(1,4))
    rdf_CaH.append(dummy[0][0])
```

```python
plt.plot(r,2*np.mean(rdf_CaO,axis=0))
plt.plot(r,np.mean(rdf_CaH,axis=0))
#plt.plot(r,np.mean(rdf_OO,axis=0))

plt.xlim([0,10])
plt.show()
```
