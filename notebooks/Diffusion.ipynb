```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from scipy.optimize import curve_fit
```

```python
def func(x, a, b):
    return a * x + b
```

```python
def func2(x, b):
    return 0. * x + b
```

# Water

```python
diff=[]
size=np.array([592,2365,4729])
path="iter-77/molecular-dynamics/production-diffusion/"
files=np.array(["ca/diff.txt","ca-med/diff.txt","ca-large/diff.txt"])
for file in files:
    data=np.genfromtxt(path + file)
    time=data[:,0]*0.0005
    msd_w=data[:,1]
    #msd_ca=data[:,2]
    popt, pcov = curve_fit(func, 6*time, msd_w)
    diff.append(popt[0])
    plt.plot(time,msd_w)
plt.show()
```

```python
plt.scatter(1/size,diff)
popt, pcov = curve_fit(func, 1/size,diff)
x = np.linspace(0,0.002)
plt.plot(x,func(x,*popt))
print(popt[1])
plt.show()
```

# Ca

```python
diff=[]
size=np.array([592,2365,4729])
path="iter-77/molecular-dynamics/production-diffusion/"
files=np.array(["ca/diff.txt","ca-med/diff.txt","ca-large/diff.txt"])
for file in files:
    data=np.genfromtxt(path + file)
    time=data[:,0]*0.0005
    #msd_w=data[:,1]
    msd_ca=data[:,2]
    popt, pcov = curve_fit(func, 6*time, msd_ca)
    diff.append(popt[0])
    plt.plot(time,msd_ca)
plt.show()
```

```python
diff=[]
size=np.array([592,2365,4729,4729,4729,4729])
path="iter-77/molecular-dynamics/production-diffusion/"
files=np.array(["ca/diff.txt","ca-med/diff.txt","ca-large/diff.txt","ca-large-2/diff.txt","ca-large-3/diff.txt","ca-large-4/diff.txt"])
for file in files:
    data=np.genfromtxt(path + file)
    time=data[:,0]*0.0005
    #msd_w=data[:,1]
    msd_ca=data[:,2]
    popt, pcov = curve_fit(func, 6*time, msd_ca)
    diff.append(popt[0])
    plt.plot(time,msd_ca)
plt.show()
```

```python
plt.scatter(1/size,diff)
print(np.mean(diff),np.std(diff)/np.sqrt(len(diff)))
popt, pcov = curve_fit(func2, 1/size,diff)
x = np.linspace(0,0.002)
plt.plot(x,func2(x,popt))
print(popt)
plt.show()
```
